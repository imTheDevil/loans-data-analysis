```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
q1 = pd.read_csv("Q1.csv", header=0)
q2 = pd.read_csv("Q2.csv", header=0)
q2b = pd.read_csv("Q2b.csv", header=0)
q3 = pd.read_csv("Q3.csv", header=0)
q4 = pd.read_csv("Q4.csv", header=0)
q5 = pd.read_csv("Q5.csv", header=0)
q6 = pd.read_csv("Q6.csv", header=0)
q7 = pd.read_csv("Q7.csv", header=0)
q8c = pd.read_csv("Q8c.csv", header=0)
q8a = pd.read_csv("Q8a.csv", header=0)
q9b = pd.read_csv("Q9b.csv", header=0)
q10 = pd.read_csv("Q10.csv", header=0)
q11 = pd.read_csv("Q11.csv", header=0)
q12b = pd.read_csv("Q12b.csv", header=0)
e1a = pd.read_csv("E1a.csv", header=0)
e1b = pd.read_csv("E1b.csv", header=0)
e1c = pd.read_csv("E1c.csv", header=0)
```

```python
#Q1
#Bar plot for "Distribution of loans across state"
fig, ax = plt.subplots(figsize = (15, 8))
ax.bar(q1['state'], q1['no_of_loans'])
ax.set_title("Distribution of loans across state")
ax.set_xlabel('State or UT')
ax.set_ylabel('Number of Loans')
ax.set_xticklabels(q1['state'], rotation=90)

for i,j in enumerate(q1['no_of_loans']):
    ax.text(i, j, str(j), ha='center', va='bottom', fontsize = 8)

plt.show()
```

```python
#Q2
#Pie chart for "Distribution of loans across amount range"
fig, ax = plt.subplots(figsize=(5,5))
#capturing patches, text and pct returned by ax.pie
patches, texts, pcts = ax.pie(x = q2b['no_of_loans'],
       labels = q2b['loan_range'],
       autopct = '%1.1f%%',
       wedgeprops = {'linewidth':.4, 'edgecolor':'white'},
       textprops = {'size':'large'}
      )
plt.setp(pcts, color='white', fontweight='bold')#styling only percentage values
ax.set_title('Distribution of loans across amounts')
plt.show()
```

```python
#Q3
#Pie chart for "Distribution of customers by age"
x = np.char.array(list(q3['age_range']))
y = np.array(list(q3['no_of_customers']))
percent = 100*y/y.sum()
fig, ax = plt.subplots(figsize=(7,7))
labels = ['{0} - {1:1.2f}%'.format(i,j) for i,j in zip(x, percent)]
patches, texts, pcts = ax.pie(x=y,
                              radius=0.8,
                              autopct = '%1.2f%%',
                              pctdistance=1.1
                             )
ax.legend(patches, labels, loc="upper right")
ax.set_title("Distribution of customers by age")
plt.show()
```

```python
xlabel = ('18 or less', '19-29', '30-49', '50-69', '70 or more')
ylabel = {
    '0-49999': (4, 842, 598, 20, 0),
    '50000-99999': (0, 2, 4, 1, 0),
    '200000-249999': (0, 0, 0, 0, 0),
}

x = np.arange(len(xlabel))  # the label locations
width = 0.2  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(10,15))

for attribute, measurement in ylabel.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of customers')
ax.set_xlabel('Age group')
ax.set_title('Distribution of loan amounts across various age groups in Tamil Nadu')
ax.set_xticks(x + width, xlabel)
ax.legend(loc='upper left', ncols=3)

plt.show()
```
