```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
q1 = pd.read_csv("Q1.csv", header=0)
q2 = pd.read_csv("Q2.csv", header=0)
q2b = pd.read_csv("Q2b.csv", header=0)
q3 = pd.read_csv("Q3.csv", header=0)
q4 = pd.read_csv("Q4.csv", header=0)
q5 = pd.read_csv("Q5.csv", header=0)
q6 = pd.read_csv("Q6.csv", header=0)
q7 = pd.read_csv("Q7.csv", header=0)
q8 = pd.read_csv("Q8.csv", header=0)
```

```python
#Q1
#Bar plot for "Distribution of loans across state"
fig, ax = plt.subplots(figsize = (15, 8))
ax.bar(q1['state'], q1['no_of_loans'])
ax.set_title("Distribution of loans across state")
ax.set_xlabel('State or UT')
ax.set_ylabel('Number of Loans')
ax.set_xticklabels(q1['state'], rotation=90)

for i,j in enumerate(q1['no_of_loans']):
    ax.text(i, j, str(j), ha='center', va='bottom', fontsize = 8)

plt.show()
```

```python
#Q2
#Pie chart for "Distribution of loans across amount range"
fig, ax = plt.subplots(figsize=(5,5))
#capturing patches, text and pct returned by ax.pie
patches, texts, pcts = ax.pie(x = q2b['no_of_loans'],
       labels = q2b['loan_range'],
       autopct = '%1.1f%%',
       wedgeprops = {'linewidth':.4, 'edgecolor':'white'},
       textprops = {'size':'large'}
      )
plt.setp(pcts, color='white', fontweight='bold')#styling only percentage values
ax.set_title('Distribution of loans across amounts')
plt.show()
```

```python
#Q3
#Pie chart for "Distribution of customers by age"
x = np.char.array(list(q3['age_range']))
y = np.array(list(q3['no_of_customers']))
percent = 100*y/y.sum()
fig, ax = plt.subplots(figsize=(7,7))
labels = ['{0} - {1:1.2f}%'.format(i,j) for i,j in zip(x, percent)]
patches, texts, pcts = ax.pie(x=y,
                              radius=0.8,
                              autopct = '%1.2f%%',
                              pctdistance=1.1
                             )
ax.legend(patches, labels, loc="upper right")
ax.set_title("Distribution of customers by age")
plt.show()
```

```python
#Q4
#Bar plot for "Distribution of loans by due dates"
fig, ax = plt.subplots(figsize = (8, 6))
ax.bar(q4['date_range'], q4['loans_due'])
ax.set_title("Distribution of loans by due dates")
ax.set_xlabel('Due Date Range')
ax.set_ylabel('Number of Loans Due')
ax.set_xticklabels(q4['date_range'], rotation=90)

for i,j in enumerate(q4['loans_due']):
    ax.text(i, j, str(j), ha='center', va='bottom', fontsize = 8)

plt.show()
```

```python
#Q5
#Line plot for "Unique customers contacted each day"
#line chart is best suitable to identify trend over time.
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(q5['day'], q5['unq_customers'], color='red', marker='o')
ax.set_title('Number of unique customers contacted between 02-12-2022 to 08-12-2022')
ax.set_xlabel('Date')
ax.set_ylabel('Number of unique customers')
ax.set_xticklabels(q5['day'], rotation=90)
ax.grid(True)
plt.show()
```

```python
#Q6
#Line plot for "Cstomers not contacted each day"
fig, ax = plt.subplots(figsize=(6,6))
ax.bar(q6['day'], q6['customers'], color='yellow')
ax.set_title('Number of customers not contacted each day')
ax.set_xlabel('Day')
ax.set_ylabel('Number of customer')
ax.set_xticklabels(q6['day'], rotation=90)
for i,j in enumerate(q6['customers']):
    ax.text(i, j, str(j), ha='center', va='bottom', fontsize = 10, c='black')
plt.show()
```
